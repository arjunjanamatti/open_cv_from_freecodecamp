```python
import pandas as pd
import os
```

```python
file_dir = 'C:/Users/Arjun Janamatti/Downloads/anomaly_scoring_data'
files_list = list(os.listdir(file_dir))
```

```python
df_0 = pd.read_csv(file_dir+'/'+files_list[0])
df_0.head()
```

```python
value_list = [val for val in df_0['value']]
```

```python
mean_list = []
for index, val in enumerate(value_list):
    mean_list.append(sum(value_list[:(index+1)])/(index+1))
mean_list
```

```python
overall_mean = (df_0['value'].sum())/len(df_0)
overall_mean
```

```python
# calculate rolling difference
n_steps = 2
def my_fun(x):
    return x.iloc[-1] - x.iloc[0]

df_0['value'].rolling(window=n_steps).apply(my_fun)
```

```python
df_0['DA'] = df_0['value'] - overall_mean
```

```python
df_0.to_csv('df_o.csv')
```
